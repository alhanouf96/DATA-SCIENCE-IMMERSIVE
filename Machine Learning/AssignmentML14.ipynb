# k-Means Clustering

Use the Seattle rain data an apply K-Means clustering to the dataset.

```python
#import the pakcages that we will need

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/daniel-dc-cd/data_science/master/module_4_ML/data/seattle_weather_1948-2017.csv')

numrows = 25549 # can be as large as 25549


#create an empty dataframe to hold values
kMeans_df = pd.DataFrame({

                             'today':[0.0] * numrows,
                             'tomorrow':[0.0] * numrows})

#sort columns for convience
seq = [
       'today',
       'tomorrow']
kMeans_df = kMeans_df.reindex(columns=seq)

```

```python
#pull values from the dataframe
    
for z in range(numrows):


    today = df.iloc[(z-1),1]
    tomorrow = df.iloc[z,1]
    
    
    kMeans_df.iat[z,0] = today
    kMeans_df.iat[z,1] = tomorrow

kMeans_df = kMeans_df.dropna() #exclude any rows with missing data
```

###  Apply K-Means Clustering 

```python
from sklearn.cluster import KMeans

x = kMeans_df.today.values
y = kMeans_df.tomorrow.values
x = x.reshape(kMeans_df.shape[0], 1)
y = y.reshape(kMeans_df.shape[0], 1)

kmeans = KMeans(n_clusters=5)
kmeans.fit(x,y)
y_kmeans = kmeans.predict(x)
```

```python
plt.scatter(x, y, c=y_kmeans, s=50, cmap='viridis')
```
