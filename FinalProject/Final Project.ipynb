# The impact of COVID-19 on Saudi stock market 


This project explains the impact of COVID-19 on the Saudi stock market during the period 2020/03/02 - 2020/10/01, As it included the stocks of various sectors (banks /basic materials /food retail /public facilities /communications and health care etc.)


Saudi Arabia COVID-19 data source:https://datasource.kapsarc.org/explore/dataset/saudi-arabia-coronavirus-disease-covid-19-situation/information/?flg=ar&disjunctive.daily_cumulative&disjunctive.indicator&disjunctive.event&disjunctive.city_en&disjunctive.region_en


As for stocks, I used pandas_datareader package to read data from yahoo website


## Import the pakcages that we will need


```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as web
```

## 1. Read Saudi Arabia COVID-19 data 

```python
COVID = pd.read_csv('Desktop/COVID-19 SA.csv')
```

## Tidies Data

```python
#Set the date as index for data

COVID=COVID.set_index('Date')
```

### Visualize active cases total

```python

COVID['Active_cases_Total'].plot(figsize=(6, 7), linewidth=3.0)
plt.style.use('dark_background')
plt.savefig('Desktop/plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Determine the highest number of cases reached

COVID['Active_cases_Total'].max()
```

## 2. Read stocks data 

will use pandas_datareader package to read data from yahoo website


```python
#Stock companies we used:

#Dr. Sulaiman Al Habib Medical Services Group Company (4013.SR)
#Abdullah Al-Othaim Markets Company (4001.SR)
#Farm Superstores (4006.SR)
#Saudi Telecom Company (7010.SR)
#Herfy Food Services Company (6002.SR)
#Alinma Bank (1150.SR)
#Qassim Cement Company (3040.SR)
#Saudi Airlines Catering Company (6004.SR)
#Saudi Electricity Company (5110.SR)


#read Stock data from 2020-02-20 to 2020/05/27

tickers = ["4013.SR", "4001.SR", "4006.SR", "7010.SR","6002.SR","1150.SR","3040.SR","6004.SR","5110.SR"]

prices = web.get_data_yahoo(tickers,
start = "2020-02-20",
end = "2020-5-27")

```

## Tidies Data

```python
#Assigning a name to the Stock instead of numbers symbol

prices=prices.rename(columns={"3040.SR" : "Qassim Cement","4001.SR":"Al-Othaim Markets", "4006.SR": "Farm Superstores", "7010.SR" : "Saudi Telecom" , "6002.SR" : "Herfy","1150.SR":"Alinma Bank","4013.SR":"Al Habib Hospital","6004.SR":"Saudi Airlines Catering","5110.SR":"Saudi Electricity"})
```

### Visualize All stocks price data

```python

x=prices["Adj Close"].plot(figsize=(6, 11),linewidth=3.0)
plt.xlabel("Date")
plt.ylabel("Adjusted Price")
plt.title("All stocks price data")
plt.style.use('dark_background')
plt.legend(title='Stocks', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('Desktop/plot2.png', dpi=300, bbox_inches='tight')
plt.show()
```

### There is a general decline in all stocks at 2020/03/10, why? 

```python
#look at COVID-19 data at 2020/03/10, this what its happen in that day

COVID.iloc[7]['Event']
```

### Visualize Stocks that grow data

```python
df = prices.xs('Adj Close', axis=1)


up= df[["Al Habib Hospital","Al-Othaim Markets","Farm Superstores","Saudi Telecom"]].plot(figsize=(5, 7),linewidth=3.0)
plt.xlabel("Date")
plt.ylabel("Adjusted Price")
plt.title("Stocks that grow")
plt.style.use('dark_background')
plt.legend(title='Stocks', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('Desktop/plot3.png', dpi=300, bbox_inches='tight')
```
