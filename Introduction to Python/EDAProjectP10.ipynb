## Assumtions based on data analysis


**#Assumptions based on the problem description noted earlier:**

1- Women (Sex=female) were more likely to have survived.

2- Children (Age<?) were more likely to have survived.

3- The upper-class passengers (Pclass=1) were more likely to have survived.



## 5. Analyze by pivoting features


```python
#Sex: there is correlation (>0.6) among Sex=female and Survived
df[['Sex', 'Survived']].groupby(['Sex'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
#Pclass: there is correlation (>0.5) among Pclass=1 and Survived

df[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

## 6. Analyze by visualizing data

```python
# how many survived, how many died ?


df['Survived'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',shadow=True)
```

```python
#Confirms Assumptions #1

sns.countplot('Sex',hue='Survived',data=df).set_title('Sex:Survived vs Dead')
plt.show()
```

```python
#Confirms Assumptions #3

sns.countplot('Pclass',hue='Survived',data=df).set_title('Pclass:Survived vs Dead')
plt.show()
```

**Observations:**

Infants (Age <=4) had high survival rate.

Oldest passengers (Age = 80) survived.

Large number of 15-25 year olds did not survive.

Most passengers are in 15-35 age range

```python
g = sns.FacetGrid(df, col='Survived')
g.map(plt.hist, 'Age', bins=20)
```

**#Observations:**

1- Pclass=3 had most passengers, and most of them did not survive. 

2- Infant passengers in Pclass=2 and Pclass=3 mostly survived. Confirms Assumptions #2

3- Most passengers in Pclass=1 survived. Confirms Assumptions #3 

```python
grid = sns.FacetGrid(df, col='Survived', row='Pclass', size=2.2, aspect=1.6)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend();
```

```python
#Embarked : Survival rate lowest for S and highest for C

sns.catplot('Embarked','Survived', kind='point', data=df);
```
